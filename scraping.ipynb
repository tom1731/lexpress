```python
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd

import re
```

# get urls

```python
url = 'https://lexpresskiosk.milibris.com/l-express/l-express/'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

links = []
for link in soup.find_all('a'):
    link = link.get("href")
    if link is not None:
        links.append(link)

r = re.compile(r".*\d{4}$")
links = list(filter(r.match, links))
links
```

# urls to df

```python
articles_dict = {
'date': [],
'article': []
}

def convert_to_dict(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    articles = []
#     for article in soup.find_all('h3', class_='ui'):
#         if len(article) > 0:
#             articles.append(article.string)

    for article in soup.find_all('h4', class_='ui'):
        if len(article) > 0:
            articles.append(article.string)

    for article in soup.find_all('p'):
        if len(article) > 0:
            articles.append(article.string)

    articles.pop(0)
    del articles[-4:]
    articles = ' '.join(articles)
    
    date = re.search(r"\d{2}\s[a-z, é]+\s\d{4}", soup.find('h2').string)
    if date is not None:
        date = date.group(0)
    
    articles_dict['date'].append(date)
    articles_dict['article'].append(articles)
```

```python
base = 'https://lexpresskiosk.milibris.com'

convert_to_dict(base)
for url in links:
    convert_to_dict(base + url)
    
df = pd.DataFrame.from_dict(articles_dict)
```

# transform data

```python
df[['day', 'month', 'year']] = df['date'].str.split(' ', expand=True)
```

```python
month_dict = {
    'janvier': '01',
    'février': '02',
    'mars': '03',
    'avril': '04',
    'mai': '05',
    'juin': '06',
    'juillet': '07',
    'août': '08',
    'septembre': '09',
    'octobre': '10',
    'novembre': '11',
    'décembre': '12'
}
```

```python
df['month'].replace(month_dict, inplace=True)
df['date'] = df['day'] + '/' + df['month'] + '/' + df['year']
df['date'] = pd.to_datetime(df['date'])
df.drop(columns = ['day', 'month', 'year'], inplace=True)
```

```python
df.to_csv('lexpress.csv', index=False)
```
